```python
# Language breakdown of all user's dotfiles
# Totals of each language
import numpy as np
import pandas as pd

def languages(language_counts):
    keys = []
    for languague_count in language_counts:
        keys.append(languague_count.keys())
    return set(np.hstack(keys))

def language_total_counts(language_counts):
    counts = {}
    for language in languages(language_counts):
        counts[language] = 0
        for count in language_counts:
            if count.get(language) != None:
                counts[language] += 1 #count.get(language, 0)
    return counts
        

counts_df = pd.DataFrame(language_total_counts(language_counts).items(), columns=['Language', 'Count'])
```

```python
%matplotlib inline
ax = counts_df.plot(x='Language', y='Count', kind='bar', figsize=(20,10),
               title='Languages used in Dotfiles by GitHub Users')
ax.set_ylabel('Number of GitHub Users')
```

```python
def isVimL(count):
    return count.get('VimL') != None
def isEmacs(count):
    return count.get('Emacs Lisp') != None or count.get('Common Lisp') != None
def isBoth(count):
    return isVimL(count) and isEmacs(count)


def vim_vs_emacs():
    counts = {'VimL': 0, 'Emacs Lisp': 0, 'Both': 0}
    for count in language_counts:
        if isBoth(count):
            counts['Both'] += 1
        elif isEmacs(count):
            counts['Emacs Lisp'] += 1
        elif isVimL(count):
            counts['VimL'] += 1
    return counts

vim_emacs_df = pd.DataFrame(vim_vs_emacs().items(), columns=['Language', 'Count'])
vim_emacs_df.plot(x='Language', y='Count', kind='bar', figsize=(15, 5))
```
